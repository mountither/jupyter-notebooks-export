        Sklearn has multiple solvers that find the minimum cost
        - ‘newton-cg’
                like gradient descent, but the second derivative is used. this makes it more sensitive to the minima
                ~ drawback : attracts saddle points, not find the suitable minima
                
        - ‘sag’
               SAG method optimizes the sum of a finite number of smooth convex functions. Like stochastic gradient (SG)  
               methods, the SAG method's iteration cost is independent of the number of terms in the sum. However, by 
               incorporating a memory of previous gradient values the SAG method achieves a faster convergence rate than 
               black-box SG methods.  
               __ It is faster than other solvers for large datasets, when both the number of samples and the number of              
               features are large.
               
               ~drawbacks: only supports l2 penalisation
               
        - ‘lbfgs'
                hessian matrix is approxiamted, saving lot of memory. stores in few vectors. suitable for small dataset
                   ~ drawback : may converge to anything. Specificty. 
        - liblinear
                supports logistic regresson and lsvm. 
                    uses coordinate descent algo. solcing coordinate directions and coord hyperplane
                    applies automatic parameter selection (l1 regular.) and good for high dimension dataset
        - saga
            The SAGA solver is a variant of SAG that also supports the non-smooth penalty=l1 option (i.e. L1 
            Regularization). This is therefore the solver of choice for sparse multinomial logistic regression and it’s also 
            suitable very Large dataset. 
                
                
                
                
                LASSO VS RIDGE (penalty)

                LASSO (the absolute value penalty)
                
                
<img src="lasso_ridge2.png">                
                
                
                
                
                
Ridge discourages large weights by setting a penalty on their squared values, which tends to drive all weights to get smaller (but not exactly zero). If there are redundant inputs, they will get a similar-sized weight.

Lasso discourages large weights by setting a penalty on their absolute values, which tends to drive all weights to get smaller, and even sets weights to exactly zero for inputs that are somewhat redundant (so you can consider removing those).

Then, you can combine the Ridge and Lasso techniques, such as with Elastic Net regularisation.

Receiver Operating Characteristic(ROC) curve is a plot of the true positive rate against the false positive rate

### accuracy 

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
acc_red1 = metrics.accuracy_score(y_test, y_pred)
```

```python
FP = cnf_matrix.sum(axis=0) - np.diag(cnf_matrix)  
FN = cnf_matrix.sum(axis=1) - np.diag(cnf_matrix)
TP = np.diag(cnf_matrix)
TN = cnf_matrix.sum() - (FP + FN + TP)
```

#### true positive rate 

```python
TPR_red1 = TP/(TP+FN)
TPR_red1 = TPR_red1.mean()
```

#### true negative rate 

```python
TNR_red1 = "N/A"
```

#### false positive rate 

```python
FPR_red1 = FP/(FP+TN)
FPR_red1 = FPR_red1.mean()
```

#### false negative rate 

```python
FNR_red1 = FN/(TP+FN)
FNR_red1 = FNR_red1.mean()
```

### percision

'micro':

Calculate metrics globally by considering each element of the label indicator matrix as a label.

'macro':

Calculate metrics for each label, and find their unweighted mean. This does not take label imbalance into account.

'weighted':

Calculate metrics for each label, and find their average, weighted by support (the number of true instances for each label).

'samples':

Calculate metrics for each instance, and find their average.


```python
print("Precision for red_wine test 1:",metrics.precision_score(y_test, y_pred, average="weighted", labels=np.unique(y_pred)))
prec_red1 =  metrics.precision_score(y_test, y_pred, average="weighted", labels=np.unique(y_pred))
```

Why do we take all the trouble doing the transformation from probability to log odds?  
-One reason is that it is usually difficult to model a variable which has restricted range, such as probability.  This transformation is an attempt to get around the restricted range problem.  It maps probability ranging between 0 and 1 to log odds ranging from negative infinity to positive infinity.
